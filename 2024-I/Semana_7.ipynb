# **Diccionarios**
 La programación se trata principalmente de resolver problemas del mundo real de la manera más eficiente posible, pero también es importante escribir y organizar el código de una manera legible para los humanos.

 En ese sentido, un **diccionario** en Python proporciona una forma más práctica de organizar los datos, es decir, funciona como una especie de "memoria asociativa" o almacenamiento de tipo "clave:valor" (*keys:values*).

 Es como tener un sistema de almacenamiento que te permite acceder rápidamente a datos utilizando identificadores únicos (claves) para recuperar información relacionada (valores). Por lo tanto, el diccionario te permite describir dos piezas de datos y la relación que poseen.

```python
# Creamos un diccionario Notas donde las claves son nombres de los alumnos
# y el valor es la nota que obtuvieron
# con llaves
Notas = {'Frank': 8,'Ronaldo': 18,'Serena': 15}
```

```python
# visualizamos los datos del diccionario Notas

Notas
```

Podemos darnos cuenta de ciertas características sobre los **diccionarios**:

* Un diccionario está rodeado por llaves **`{}`** y los pares **"clave:valor"** están separados por comas.
* Un diccionario consta de uno o más pares "**clave:valor**". La clave es el "identificador" o "nombre" que se utiliza para describir el valor. Por ejemplo, en el diccionario 'diccionario_notas', uno de los pares **"clave:valor"** es 'Frank' : 8
* Las **claves** en un diccionario son únicas.
* La sintaxis para un par clave/valor es **"CLAVE: VALOR"**.
* Las **claves** (por ejemplo, 'Frank') en un diccionario deben ser **inmutables**.
* Los **valores** (por ejemplo, 8) en un diccionario pueden ser **cualquier objeto de Python**.
* Un diccionario puede estar vacío.

Veamos qué tipos de datos funcionan y/o sirven como "claves" en un diccionaro. Por ejemplo, tenemos a los strings, los cuales sí funcionan como claves.

```python
### Los valores pueden ser listas

Notas_finales = {'Frank': [8, 10, 15], 'Susan': [7, 5, 12], 'Guido': [20, 18, 19]}
```

```python
Notas_finales
```

```python
### cómo obtener los keys de un diccionario
## .keys()
Notas_finales.keys()
```

```python
## cómo obtener los values de cada key de un diccionario
## .values()
Notas_finales.values()
```

No obstante, en el caso de las listas, estas no funcionan como claves en un diccionario.

```python
diccionario_incorrecto = {['a', 'list']: 8}
```

## **¿Cómo agregar elementos a un diccionario?**
Hay una manera muy simple de agregar un par "**clave:valor**" a un diccionario. Por favor, observa el siguiente fragmento de código:


```python
## Crear un diccionario vacio

a_dict = dict()
print(a_dict)
```

```python
## Agregamos cada key y su valor

## La forma es nombre_diccionario["clave"] = valor
## Recuerda la clave o key siempre es un string
### Asimismo, el valor o vlaue puede ser todo tipo de datos y listas

a_dict['Frank'] = 8
print(a_dict)
```

Ten en cuenta que las claves del diccionario deben ser identificadores **únicos** para los valores en el diccionario. En ese sentido, los pares "**clave:valor**" se sobrescriben si asignas un valor diferente a una clave existente.


```python
## Creamos un diccionario vacio
b_dict = dict()

b_dict['Frank'] = 8
print(b_dict)
b_dict['Frank'] = 7
print(b_dict)
b_dict['Frank'] = 9
print(b_dict)

###Solo el último valor asociado al key "Frank" se conserva
### Por lo tanto, los valores de un key pueden cambiarse
```

```python
## Creamos un diccionario vacio y agregamos cada clave y valor
c_dict = dict()

### Agregamos una clave con su valor
c_dict["Frank"] = 8

## Agregamos una segunda clave con su valor
c_dict["Luisa"] = "peruana"

## Agregamos una tercer clave con una lista como valor
c_dict["Angie"] = [28, "argentina"]

# Visualizamos
print(c_dict)
```

## **Contar con un diccionario**
Los diccionarios son muy útiles para derivar estadísticas. Por ejemplo, podemos determinar fácilmente la frecuencia de cada letra en una palabra.


```python
## Creamos un diccionario vacio
frecuencia_letras = dict()

### Creamos una variable que contiene un string
word = 'hipohipohi'

## Recorremos un bucle for donde cada letra se convierte en una clave
for letra in word:
    if letra in frecuencia_letras: # se agrega 1 al diccionario si la clave existe
        frecuencia_letras[letra] = frecuencia_letras[letra] + 1 # nota: x +=1 hace lo mismo que x = x + 1
    else:
        frecuencia_letras[letra] = 1 # establecermos un valor predeterminado de 1 cuando no hay más repeticiones
print(frecuencia_letras)
```

```python
## Creamos una lista de palabras
frase = ['Obama', 'fue', 'un', 'presidente', 'de', 'los', 'EE. UU', "Castillo", "fue", "un", "presidente", "del", "Perú"]

## Creamos un diccionario vacio
frecuencia_palabra = dict()

## Recorremos un bucle for donde cada palabra se convierte en una clave
for palabra in frase:

    if palabra in frecuencia_palabra: # agregar 1 al diccionario si la clave existe
        frecuencia_palabra[palabra] = frecuencia_palabra[palabra] + 1
    else:
        frecuencia_palabra[palabra] = 1 # establecer un valor predeterminado de 1 cuando no hay más repeticiones

print(frecuencia_palabra)
```

## **Diagramas de barras verticales**
`bar(x, y)`: Dibuja un diagrama de barras verticales donde x es una lista con la posición de las barras en el eje X, e y es una lista con la altura de las barras en el eje Y.

```python
import matplotlib.pyplot as plt

# Extraer las palabras y sus frecuencias
clave_palabra = list(frecuencias.keys())
veces_palabra = list(frecuencias.values())

# Crear el gráfico de barras
plt.figure(figsize=(100, 10))  # Ajusta el tamaño del gráfico si es necesario
plt.bar(clave_palabra, veces_palabra, color='orange')  # Puedes cambiar el color

# Añadir etiquetas y títulos
plt.ylabel('Frecuencia')
plt.xlabel('Palabras')
plt.title('Frecuencia de Palabras en la Lista')


# Rota las etiquetas del eje x para que se vean mejor
plt.xticks(rotation=90)

# Establecer los límites del eje x , esto permite que se elimine el espacio en blanco que se tenía antes del primer elemento del eje de x
plt.xlim(left=0)


# Guardamos
plt.savefig('ejemplo_1a.jpg', format='jpg')

plt.show()
```

# **Ejercicio 2:**

Crea un algoritmo que determine quiénes podrían ser fanáticos de Taylor Swift, basándote en una serie de preguntas cuyas respuestas indiquen la probabilidad de que una persona sea un fanático.

Usa bucle while con varias condiciones if, elif, y else para procesar las respuestas de los usuarios y clasificar su nivel de fanatismo.

```python
## Iniciamos un progrma que va determinar si el usuario es o no fanático de Taylor Swift
print("Responde las siguientes preguntas para saber si eres fanático de Taylor Swift:")

#La variable puntos se inicia en 0. Esta variable se usará para acumular puntos basados en las respuestas del usuario.
puntos = 0

# El programa entra en un bucle infinito que continuará hasta que se rompa explícitamente con un break.
while True:
    # Pregunta 1
    respuesta = input("¿Te gusta la música pop o country? (sí/no): ").lower()
    if respuesta == "sí":
        puntos = puntos + 1  # Suma puntos si le gusta la música pop o country
    elif respuesta == "no":
        puntos = puntos - 1 # Se resta puntos si no le gusta la música pop o country
    else:
        print("Respuesta no válida. Por favor responde 'sí' o 'no'.")
        continue

    # Pregunta 2
    respuesta = input("¿Puedes nombrar al menos tres álbumes de Taylor Swift? (sí/no): ").lower()
    if respuesta == "sí":
        puntos = puntos + 2  # Suma más puntos si conoce bien su discografía
    elif respuesta == "no":
        puntos = puntos - 1  # Se resta puntos si no conoce bien su discografía
    else:
        print("Respuesta no válida. Por favor responde 'sí' o 'no'.")
        continue

    # Pregunta 3
    respuesta = input("¿Has ido alguna vez a un concierto de Taylor Swift? (sí/no): ").lower()
    if respuesta == "sí":
        puntos = puntos + 2  # Suma puntos adicionales si ha asistido a conciertos
    elif respuesta == "no":
        puntos = puntos + 0  # No suma ni resta si no ha ido a conciertos
    else:
        print("Respuesta no válida. Por favor responde 'sí' o 'no'.")
        continue

    break  # Rompe el bucle una vez que todas las preguntas han sido respondidas

# Evaluación final basada en la puntuación
if puntos > 3:
    print("¡Eres un gran fanático de Taylor Swift!")
elif puntos > 0:
    print("Parece que te gusta Taylor Swift, pero quizás no te consideras un fanático.")
else:
    print("No eres un fanático de Taylor Swift")
```
