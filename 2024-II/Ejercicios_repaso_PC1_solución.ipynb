## **Ejercicio 1**

Escribe un programa que interactúe con el usuario para recolectar la siguiente información:

Primero, solicita el **título de la noticia**. A continuación, pide al usuario que indique la **cantidad de palabras** que contiene la nota. Luego, pregunta por la **fuente de la noticia**, es decir, el sitio web o medio del cual proviene. Finalmente, **consulta si la noticia es urgente o no**.

El programa debe realizar las siguientes validaciones:

Primero, **verifica si el título de la noticia tiene más de 20 caracteres**. Luego, **asegúrate de que la cantidad de palabras esté en un rango válido**, es decir, **entre 300 y 2000 palabras**. A continuación, **revisa que la fuente proporcionada finalice en ".com" o ".org" para validar que proviene de un medio confiable**. Finalmente, **si la noticia es clasificada como urgente (es decir, el título comienza con la palabra "URGENTE"), el programa deberá imprimir un mensaje adicional: "¡Noticia urgente detectada!"**

----
## Solución

```python
# Creamos una variable donde el usuario asignará el título de la noticia
título_noticia = input("Ingrese el título de la noticia: ")
# Creamos otra variable para que el usuario ingrese la cantidad de palabras de su noticia
# También convertimos el dato ingresado a enteros
cantidad_palabras = int(input("Ingrese la cantidad de palabra de su noticia: "))
# Creamos otra variable donde el usuario debe ingresar la fuente de la noticia
fuente_noticia = input("Ingrese la fuente de la noticia: ")

# Ahora verificamos los datos ingresados con declaraciones condicionales
if len(título_noticia) > 20: # Verificamos que cumpla con la cantidad de caracteres el título
   if 300 < cantidad_palabras < 2000:  # Verificamos que la cantidad de palabras esté en el rango
      if fuente_noticia.endswith(".com") or fuente_noticia.endswith(".org"): # Verificamos que la fuente sea confiable
         if "URGENTE" in título_noticia:  # Verificamos si la noticia es URGENTE
            print("¡Noticia urgente detectada!")
         else:
             print("Noticia no urgente.")
      else:
          print("La noticia no tiene una fuente confiable.")
   else:
       print("La noticia no tiene la cantidad de palabras requerida.")
else:
    print("El título es muy corto.")
```

## **Ejercicio 2**
Escribe un programa que recoja la siguiente información del usuario:

Primero, solicita la **duración del video en segundos**. Luego, pide al usuario que indique el **formato del archivo de video**. Posteriormente, pregunta por la **resolución del video en el formato anchoxalto** (por ejemplo: 1920x1080). **El programa deberá verificar si el video es compatible con redes sociales y mostrar un mensaje especial en caso de que lo sea**.

El programa debe cumplir con las siguientes validaciones:

Primero, **verifica que la duración del video esté entre 30 segundos y 2 horas (7200 segundos)**. A continuación, **asegúrate de que el formato del archivo sea uno de los aceptados: .mp4, .mov o .avi**. Después, **verifica que la resolución del video sea al menos 1920x1080**. Finalmente, **si el video tiene una duración de menos de 1 minuto, muestra un mensaje indicando que es compatible con redes sociales**.

----
## Solución

```python
# Primero, creamos solicitamos la duración del video (en segundos)
duración_video = input("Ingrese la duración del video en segundos: ")

# Creamos una variable que solicite al usuario el formato del video
formato_video = input("Ingrese el formato del video (.mp4, .mov o .wav): ")

# Creamos dos variables donde se solicite el ingreso del ancho y largo del video
ancho_video = input("Ingrese el ancho del video: ")
largo_video = input("Ingrese el largo del video: ")

# Mediante declaraciones condicionales verificaremos que se haya cumplido con los requerimientos
# Durante el proceso de verificación realizo la conversión a enteros de los caracteres numéricos ingresados por el usuario
if 30 < int(duración_video) < 7200: # Verificamos que la duración esté en el rango indicado
   if formato_video.endswith(".mp4") or formato_video.endswith(".mov") or formato_video.endswith(".wav"): # Verificamos que el formato se cumpla
      if int(ancho_video) >= 1920: # Verificamos que el ancho cumpla el mínimo especificado
         if int(largo_video) >= 1080: # Verificamos que el largo cumpla el mínimo especificado
            if int(duración_video) < 60: # Verificamos que la duración ingresada sea la máxima permitida para ser publicada en redes sociales
               print("El video es compatible para ser publicado en las redes sociales.")
            else:
               print("El video no es compatible para ser publicado en las redes sociales.")
         else:
             print("El largo del video no es compatible.")
      else:
         print("El ancho del video no es compatible.")
   else:
      print("El formato del video no es compatible.")
else:
   print("La duración del video no es compatible.")
```

## **Ejercicio 4**

Escribe un programa en Python que busque una palabra específica en una frase famosa de la película El Padrino.

Define la frase de El Padrino como:
**`"Voy a hacerle una oferta que no podrá rechazar"`**.

Utiliza el método `split()` para dividir la frase en una lista de palabras.

**Establece una variable cuyo valor sea "oferta".**

Accede a las palabras de la lista individualmente utilizando su posicionamiento (índice) y emplea una estructura condicional if-elif-else para realizar lo siguiente:

- **If** la palabra en la primera posición (índice 0) es "oferta", imprime un mensaje indicando su posición.
- **Elif** la palabra en la segunda posición (índice 1) es "oferta", imprime un mensaje indicando su posición.
- **Repite este proceso para las siguientes posiciones.**
- **Else:** si la palabra "oferta" no se encuentra en ninguna posición, imprime un mensaje indicando que no fue encontrada.

```python
# Creamos una variable donde le asignamos la frase de la película El padrino
frase = "Voy a hacerle una oferta que no podrá rechazar"

# Convertimos la anterior variable en una lista
lista_frase = frase.split()

# Creamos otra variable cuyo valor sea "oferta"
palabra_buscada = "oferta"

# Verificamos cada elemento de la lista frase con la palabra buscada
if lista_frase[0] == palabra_buscada:
   print("La palabra 'oferta' se ubica en la primera posición de la lista.")
elif lista_frase[1] == palabra_buscada:
   print("La palabra 'oferta' se ubica en la segunda posición de la lista.")
elif lista_frase[2] == palabra_buscada:
   print("La palabra 'oferta' se ubica en la tercera posición de la lista.")
elif lista_frase[3] == palabra_buscada:
   print("La palabra 'oferta' se ubica en la cuarta posición de la lista.")
elif lista_frase[4] == palabra_buscada:
   print("La palabra 'oferta' se ubica en la quinta posición de la lista.")
elif lista_frase[5] == palabra_buscada:
   print("La palabra 'oferta' se ubica en la sexta posición de la lista.")
elif lista_frase[6] == palabra_buscada:
   print("La palabra 'oferta' se ubica en la séptima posición de la lista.")
elif lista_frase[7] == palabra_buscada:
   print("La palabra 'oferta' se ubica en la octava posición de la lista.")
elif lista_frase[8] == palabra_buscada:
   print("La palabra 'oferta' se ubica en la novena posición de la lista.")
else:
  print("La palabra 'oferta' no se encuentra en la lista.")
```
