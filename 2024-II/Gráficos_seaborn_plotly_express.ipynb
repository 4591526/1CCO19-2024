## **1. Seaborn**

- Estilo de gráficos atractivo: Seaborn aplica estilos predeterminados que mejoran la apariencia visual de los gráficos, haciendo que luzcan más profesionales y fáciles de interpretar

- Facilidad para crear gráficos estadísticos: Incluye gráficos de alta complejidad estadística, como diagramas de dispersión con regresiones lineales (regplot), histogramas con distribuciones de densidad (distplot), entre otros.

- Paletas de colores personalizadas: La librería incluye una amplia gama de paletas de colores predefinidas (como deep, muted, bright, dark, colorblind) y la posibilidad de crear paletas personalizadas para adaptar los gráficos a diferentes propósitos o estilos.

- Combinación con Matplotlib: Aunque es una librería independiente, Seaborn permite ajustar y personalizar gráficos aún más usando funciones de Matplotlib, ampliando las posibilidades de personalización.

````python
# Importamos las librerías (solo es necesario importar una vez)
import pandas as pd     # Te permite trabajar con archivos csv o excel
import matplotlib.pyplot as plt  # Te prmite hacer gráficos 2d
````

````python
# Cargamos el archivo de Excel en un DataFrame
ligue_1 = pd.read_excel('F1-2023-2024.xlsx')

# Lee el archivo CSV (asegúrate de que el nombre del archivo sea correcto)
df_csv = pd.read_csv("F1-2023-2024.csv")
````

````python
# Visualizamos la tabla
print(ligue_1.head())
````

````python
# Visualizamos la tabla
df_csv.head()
````

````python
# Calculamos el promedio de goles anotados como local para cada equipo
# groupby() agrupa las filas de un DataFrame según los valores de una columna
# mean() calcula el promedio de los valores de una columna para cada grupo
promedio_goles_por_equipo = df_csv.groupby('HomeTeam')['FTHG'].mean()

# Esta parte del código agrupa el DataFrame df_csv por la columna 'HomeTeam'.
# Agrupar por 'HomeTeam' significa que se crearán grupos de filas donde cada grupo
# contiene todos los partidos jugados por el mismo equipo como local.
````

````python
# Visualizamos los promedios
print(promedio_goles_por_equipo)
````

````python
# Averiguamos el tipo de dato que es la variable promedio_goles_por_equipo
print(type(promedio_goles_por_equipo))
````

````python
# Convierte la Serie en DataFrame con la función .reset_index()
promedio_goles_por_equipo_df = promedio_goles_por_equipo.reset_index()

# Visualizamos
promedio_goles_por_equipo_df
````

````python
# Importamos la librería Seaborn
import seaborn as sns

# Creamos las dimensiones del gráfico con matplotlib
plt.figure(figsize=(19, 8))

# Configuración del gráfico
sns.barplot(
    x='HomeTeam',
    y='FTHG',
    hue='HomeTeam',  # Se utiliza hue para que Seaborn asigne la paleta de colores de acuerdo a cada equipo
    data=promedio_goles_por_equipo_df,
    palette='Set2',
    legend=False  # Oculta la leyenda ya que solo hay un valor `hue` por equipo
)

# Añade etiquetas a los ejes
plt.xlabel("Equipo Local")
plt.ylabel("Promedio de Goles como Local")
plt.title("Promedio de Goles como Local por Equipo")
plt.yticks(ha='center')

# Añade una cuadrícula en el eje y
plt.grid(axis='y', linestyle='--', alpha=1)

# Guarda el gráfico en un archivo
plt.savefig('promedio_goles.png', format='png', dpi=300, bbox_inches='tight')

# Mostrar el gráfico
plt.show()

````

---

En Seaborn, el parámetro hue es utilizado para dividir o categorizar los datos dentro de un gráfico según una variable específica, aplicando diferentes colores a cada categoría. Esto permite visualizar de manera clara cómo varían los datos entre diferentes subgrupos.

# Gráficos interactivos con Plotly express:

- Gráficos interactivos: Los gráficos creados son interactivos por defecto, lo que permite hacer zoom, desplazarse y ver detalles al pasar el cursor sobre los puntos de datos.
- Simplicidad: Requiere pocas líneas de código para generar gráficos complejos y atractivos.
- Compatibilidad con pandas: Trabaja directamente con DataFrames, lo que facilita la manipulación y el análisis de datos en conjunto con pandas.
- Variedad de gráficos: plotly.express permite crear muchos tipos de gráficos, como gráficos de dispersión, histogramas, gráficos de barras, mapas de calor, líneas de tiempo, gráficos 3D, entre otros.

````python
# Importamos la librería
import plotly.express as px

# Convierte la Serie en DataFrame
# promedio_goles_por_equipo_df.columns = ['HomeTeam', 'FTHG']  # Renombra columnas si es necesario

# Crear el gráfico de barras con Plotly Express
fig = px.bar(promedio_goles_por_equipo_df,
             x='FTHG',
             y='HomeTeam',
             orientation='h',
             title='Promedio de Goles como Local por Equipo',
             labels={'FTHG': 'Promedio de Goles', 'HomeTeam': 'Equipo Local'},
             color='HomeTeam',  # Si deseas que cada equipo tenga un color único
             color_discrete_sequence=px.colors.qualitative.Antique)

# Actualizar el diseño del gráfico
fig.update_layout(xaxis_title='Promedio de Goles', yaxis_title='Equipo Local')

# Mostrar el gráfico
fig.show()
````

## **Histogramas**

````python
### Ahora solo deseamos filtrar datos del "Paris SG"

# Filtrar los partidos donde "Paris SG" es el equipo local
partidos_local = df_csv[df_csv['HomeTeam'] == "Paris SG"]

# df_csv['HomeTeam']
# df_csv[df_csv['HomeTeam'] == "Paris SG"]

# Seleccionar las columnas relevantes para el análisis
resultados_local = partidos_local[['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR']]

# Mostrar los resultados de los partidos como local
print("Partidos como Local:")
print(resultados_local)
````

````python
# Filtrar los partidos donde "Paris SG" es el equipo visitante
partidos_visitante = df_csv[df_csv['AwayTeam'] == "Paris SG"]

# Seleccionar las columnas relevantes para el análisis
resultados_visitante = partidos_visitante[['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR']]

# Mostrar los resultados de los partidos como visitante
print("Partidos como Visitante:")
print(resultados_visitante)
````

````python
# Importamos las librerías necesarias
#import seaborn as sns
#import matplotlib.pyplot as plt

# Creamos listas de goles anotados y recibidos por "Paris SG" como local y visitante
goles_anotados_local = resultados_local['FTHG']
goles_anotados_visitante = resultados_visitante['FTAG']

# Configuración del tamaño general de las figuras
sns.set(rc={'figure.figsize': (12, 8)})

# Gráfico de distribución de goles anotados como local
sns.displot(goles_anotados_local, bins=range(0, 8), color='#3ee84e', edgecolor='black', alpha=0.75)
plt.title('PSG: goles anotados como Local')
plt.xlabel('Goles como local')
plt.ylabel('# partidos')
plt.savefig('goles_local_psg.png', dpi=300)  # Guarda solo esta parte

# Gráfico de distribución de goles anotados como visitante
sns.displot(goles_anotados_visitante, bins=range(0, 8), color='#9deda5', edgecolor='black', alpha=0.75)
plt.title('PSG: goles anotados como Visitante')
plt.xlabel('Goles como visitante')
plt.ylabel('# partidos')
plt.savefig('goles_visitante_psg.png', dpi=300)  # Guarda solo esta parte

plt.show()
````

````python
# Importamos las librerías necesarias
# import plotly.express as px
import plotly.subplots as sp # permite crear figuras con múltiples gráficos

# Crear una figura de subplots 2x2 para los histogramas
fig = sp.make_subplots(rows=1, cols=2, subplot_titles=("PSG: goles anotados como Local", "PSG: goles anotados como Visitante"))

# Histograma de goles anotados como local
fig_local = px.histogram(goles_anotados_local, nbins=10, title="PSG: goles anotados como Local", labels={'value': 'Goles como local', 'count': '# partidos'})
fig_local.update_traces(marker_color='skyblue', marker_line_color='black', opacity=0.75)

# Histograma de goles anotados como visitante
fig_visitante = px.histogram(goles_anotados_visitante, nbins=10, title="PSG: goles anotados como Visitante", labels={'value': 'Goles como visitante', 'count': '# partidos'})
fig_visitante.update_traces(marker_color='green', marker_line_color='black', opacity=0.75)

# Añadir los gráficos a la figura de subplots
for trace in fig_local.data:
    fig.add_trace(trace, row=1, col=1)

for trace in fig_visitante.data:
    fig.add_trace(trace, row=1, col=2)

# Actualizar los ejes y diseño general
fig.update_layout(title_text="Goles anotados por PSG como Local y Visitante", showlegend=False)
fig.update_xaxes(title_text="Goles", row=1, col=1)
fig.update_yaxes(title_text="# partidos", row=1, col=1)
fig.update_xaxes(title_text="Goles", row=1, col=2)
fig.update_yaxes(title_text="# partidos", row=1, col=2)

# Mostrar el gráfico
fig.show()
````

## **Gráfico de Pastel**

````python
# Filtramos los partidos donde "Paris SG" es el equipo local o visitante
partidos_psg = ligue_1[(ligue_1['HomeTeam'] == "Paris SG")]

# Creamos las variables para contar los partidos ganados, perdidos y empatados
ganados = 0
perdidos = 0
empatados = 0

# Iteramos sobre los partidos del PSG para contar los resultados y almacenarlos en las variables
# iterrows() es un método que permite iterar sobre las filas de un DataFrame
# index es el índice de la fila y row es la fila en sí
for index, row in partidos_psg.iterrows():
    if row['HomeTeam'] == "Paris SG": # Si el PSG es el equipo local en el partido
        if row['FTR'] == 'H': # Si el PSG ganó el partido
            ganados = ganados + 1 # Incrementamos el contador de partidos ganados
        elif row['FTR'] == 'A': # Si el PSG perdió el partido
            perdidos = perdidos + 1 # Incrementamos el contador de partidos perdidos
        else: # Si el partido terminó en empate
            empatados = empatados + 1 # Incrementamos el contador de partidos empatados
````

````python
print(partidos_psg.head())
````

````python
print(ganados)
print(perdidos)
print(empatados)
````

````python
# Importamos las librerías necesarias
#import seaborn as sns
#import matplotlib.pyplot as plt

# Creamos las etiquetas, tamaños y colores para el gráfico de pastel
labels = ['Ganados', 'Perdidos', 'Empatados']
sizes = [ganados, perdidos, empatados]
colors = sns.color_palette("Set2", 3)  # Definimos una paleta de colores con Seaborn

# Creamos el gráfico de pastel con Seaborn y Matplotlib
plt.figure(figsize=(8, 8))  # Tamaño del gráfico
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', explode=(0.01, 0.01, 0.01), shadow=True, startangle=140)

# Añadimos un título y ajustamos el gráfico
plt.title('Resultados de los Partidos como Local del PSG (2023-2024)')
plt.axis('equal')  # Para que el gráfico sea un círculo perfecto

# Guardamos el gráfico
plt.savefig('pastel_psg.png', dpi=300)

# Mostramos el gráfico
plt.show()
````

````python
# Importamos las librerías necesarias
# import plotly.express as px

# Creamos un DataFrame con los datos
data = pd.DataFrame({
    'Resultado': ['Ganados', 'Perdidos', 'Empatados'],
    'Cantidad': [ganados, perdidos, empatados]
})

# Creamos el gráfico de pastel con Plotly Express
fig = px.pie(data, values='Cantidad', names='Resultado',
             title='Resultados de los Partidos como Local del PSG (2023-2024)',
             color_discrete_sequence=px.colors.qualitative.Set2)  # Usamos la paleta "Set2" para colores similares

# Ajustes adicionales
fig.update_traces(hole=0.1, pull=[0.01, 0.01, 0.01], textinfo='percent+label')  # "pull" separa cada sección, "textinfo" añade etiquetas y porcentajes
fig.update_layout(showlegend=True)  # Mostrar la leyenda

# Mostrar el gráfico
fig.show()
````
