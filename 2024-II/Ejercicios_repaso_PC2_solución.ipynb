## **Ejercicio 1:**

Crea un programa que permita almacenar nombres y números de celular en un diccionario. Luego, deberán utilizar un bucle para permitir al usuario agregar múltiples contactos hasta que decidan detenerse.

- Crea un diccionario vacío para almacenar nombres como claves y números de teléfono como valores.
- **Usa un bucle while para agregar contactos, pidiendo al usuario que ingrese el nombre y el número de celular. El bucle debe continuar hasta que el usuario decida detenerse**.

```python
# Creamos un diccionario vacío para almacenar contactos
contactos = {}

# Con un bucle while, ingreamos el nombre del contacto como keys con su número como values
while True:
    # Solicitamos nombre del contacto
    nombre = input("Ingresa el nombre del contacto: ")

    # Solicitamos número de celular
    numero = input("Ingresa el número de celular del contacto: ")

    # Agregamos el contacto al diccionario
    contactos[nombre] = numero

    # Preguntamos si se desea continuar
    continuar = input("¿Deseas añadir otro contacto? (sí/no): ").lower()

    if continuar != "sí":
        break

# Mostrar el diccionario actualizado con los contactos
print(contactos)
```

## **Ejercicio 2:**

A partir del archivo RPP-octubre-2024.txt (carpeta semana 7), elabora un programa en Python que lea el contenido y elimine los saltos de línea, signos de puntuación, caracteres numéricos, y convierte todo a minúsculas.

Luego, almacena cada palabra del texto en un diccionario donde las claves son las palabras y los valores son las veces que aparece cada una.

Finalmente, utiliza la librería WordCloud para crear la visualización y guarda el gráfico con un nombre representativo.

Referencias:
RPP Noticias. (2024, octubre 3). Un millón de viviendas quedarían destruidas o inhabitables en Lima y Callao con sismo de magnitud 8.8, advierte experta. https://rpp.pe/lima/actualidad/un-millon-de-viviendas-quedarian-destruidas-o-inhabitables-en-lima-y-callao-con-sismo-de-magnitud-88-advierte-experta-noticia-1588939

```python
# El siguiente código nos permite abrir un archivo del escritorio
from google.colab import files
uploaded = files.upload()
```

```python
# Leemos el archivo seleccionado y le asignamos una variable
with open("RPP-octubre-2024.txt") as file:
    noticia = file.read()

# Visualizamos
print(noticia)
```

```python
# Convertimos el texto a minúsculas
texto_min = noticia.lower()

# Eliminamos los signos de puntuación
texto_depurado = texto_min.replace(",", "").replace(".", "").replace(";", "").replace("¿", "").replace("?", "").replace('"', "").replace("(", "").replace(")", "").replace(":", "")

# Visualizamos
print(texto_depurado)
```

```python
# Creamos una lista de palabras con el contenido de la depuración de comentarios
palabra_noticia = texto_depurado.split()

print(palabra_noticia)
```

```python
# Creamos un diccionario vacío
noticia_rpp = dict()

# El bucle for recorre las palabras de la lista original (paro)
for elemento in palabra_noticia:
        if elemento in noticia_rpp: # Si la palabra aparece más de una vez, se suma una unidad al valor (frecuencia)
            noticia_rpp[elemento] = noticia_rpp[elemento] + 1
        else:
            noticia_rpp[elemento] = 1 # Si solo aparece una vez, su valor será solo uno

print(noticia_rpp)
```

## Nube de palabras

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generamos la nube de palabras
wordcloud = WordCloud(width=900, height=670, background_color="white", colormap='viridis').generate_from_frequencies(noticia_rpp)

# Mostramos la nube de palabras
plt.figure(figsize=(4,4))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

# Guardamos la nube de palabras en un archivo
wordcloud.to_file("noticia_rpp.png")
```

# Gráfico de barras verticales

```python
# Importamos la librería matplotlib que nos permitirá realizar gráficos de barras
# import matplotlib.pyplot as plt

# Extraemos las palabras y sus frecuencias
clave_palabra = list(noticia_rpp.keys())
veces_palabra = list(noticia_rpp.values())

# Creamos el gráfico de barras
plt.figure(figsize=(25, 8))
plt.bar(clave_palabra, veces_palabra, color='orange')  # indicamos que las barras sean verticales

# Añadimos etiquetas y títulos
plt.ylabel('Frecuencias')
plt.xlabel('Palabras')
plt.title('Frecuencia de las palabras de la noticia RPP')

# Rotamos las etiquetas del eje x para que se vean mejor
plt.xticks(rotation=90)

# Establecemos los límites del eje x , esto permite que se elimine el espacio en blanco que se tenía antes del primer elemento del eje de x
plt.xlim(left=-0.5)

# Mostramos
plt.show()

# Guardamos
plt.savefig('rpp_barras_verticales.png', format='png')
```

Se proporcionará un diccionario llamado **cartelera** con **categorías específicas para diferentes géneros de películas (claves)**, **además de una categoría "Otros" para aquellos géneros no predefinidos**.
