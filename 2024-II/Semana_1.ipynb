## **2. "Hola mundo" en Python**
Al aprender a programar, se acostumbra comenzar con un ejemplo sencillo: `print("Hola mundo")`. Esta línea de código asegurará que sepamos cómo imprimir una cadena de caracteres (string) en el output (salida) y cómo ejecutar el código dentro de las celdas en un notebook.

```python
# Probemos el primer output en Python

print('Hola mundo')
```

### **¿Qué está ocurriendo aquí?**
Estamos usando la función `print()` para imprimir el texto `'Hola mundo'` en la pantalla (output). Es decir, la función `print()` muestra la cadena de caracteres (argumento)dentro de sus paréntesis en la pantalla.

**Nota:** Una función es un bloque de código que realiza una tarea específica. Las funciones toman entradas (llamadas argumentos, un término para referise a los valores que pones dentro de una función) y devuelven salidas. En este caso, la función `print()` toma el texto `'Hola mundo'` como entrada y devuelve el mismo texto como salida, que se muestra en la pantalla.

**¿Cómo podemos modificar el código para que muestre tu nombre en lugar de 'Hola mundo'?**

```python
# Imprimamos nuestro nombre

print("Luisa")
```

## **3. Escribir comentarios en Python**
¿Viste que usamos el símbolo # para añadir un comentario al código? Los comentarios son líneas de texto que no se ejecutan como código, sino que se utilizan para hacer que el código sea más fácil de entender. Puedes añadir comentarios a tu código para recordar lo que hace, o para ayudar a otros a entenderlo.

Para escribir comentarios en Python, use el símbolo numeral `#` antes de escribir tu comentario. Cuando ejecutes tu código, Python ignorará todo lo que esté más allá del `#`.

```python
# Escribamos un comentario

print('Facultad de Ciencias y Artes de la Comunicación') # Esta línea imprime un string
# print('Hola')
```

## **5. ¿Python sabe acerca de tu error antes de ejecutar tu código?**
Python es lo que se llama un lenguaje interpretado. Por su parte, los lenguajes compilados (C++, Java, entre otros) examinan todo el programa en el momento de la compilación y pueden advertirle sobre toda clase de errores antes de la ejecución. Por el contrario, Python interpreta línea por línea a medida que lo ejecuta. Python dejará de ejecutar todo el programa cuando encuentre un error.

```python
# Recorremos los siguientes strings para ver el orden de ejecución

print("Esto se ejecutará")
frint("Esto tendrá un error")
print("Esto NO se ejecutará")
```

### **Función `print()`**
* La función `print()` muestra el valor de una cadena de caracteres (string) dentro de los paréntesis en la pantalla (output).
* La función `print()` admite varios argumentos seguidos y separados por comas.

Los siguientes argumentos se muestran en el mismo orden y en la misma línea, separados por un espacio:

```python
print("Hola","Adiós") # hay un espacio
```

```python
print("Hola"," Adiós") # hay dos epacios
```

```python
print("Hola ","Adiós") # hay dos espacios
```

```python
print("Hola""Adiós") # no hay espacio
```

Cuando se trata de dos cadenas seguidas, se puede no escribir comas entre ellas. Sin embargo, las cadenas (strings) se escribirán sin un espacio en blanco entre ellas:

```python
print("Hola" "Adiós") # es igual a print("HolaAdiós")
```

También se puede ejecutar dos argumentos y especificar un separador.

```python
print("Hola", "¿Cómo están?", sep="---")
```

```python
print("Hola", "¿Cómo están?", sep=";")
```

```python
print("Hola", "¿Cómo están?", sep=":")
```

```python
print("Hola", "¿Cómo están?", sep=": ")
```
