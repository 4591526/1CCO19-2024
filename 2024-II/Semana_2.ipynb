## **1. Variables**
En vez de escribir el valor directamente en la función `print()`, podemos **asignar** el valor a una variable y luego usar la variable dentro de la función `print()`.

Cuando pasamos un valor a la función `print()`, obtenemos el mismo resultado que si pasamos una variable que contiene ese valor.



```python
print("Hola mundo")
```

```python
frase = "Hola mundo"

print(frase)
```

```python
perro = "Guau"
print(perro)
```

```python
Guau = "sonido de un perro"
print(Guau)
```

"Guau" es un pedazo de texto que llamamos una cadena de caracteres o string en inglés.

* Los strings en Python siempre van entre comillas simples o dobles. Recuerden que sin esas comillas, Python pensará que Guau es una variable porque los nombres de las variables no van entre comillas.

* Los lenguajes de programación pueden verse como maneras 'formales' de decirle a tu computadora que tiene que hacer. Por lo tanto, los lenguajes de programación hacen uso de reglas estrictas y bien definidas (conocidas como sintaxis).

* Más adelante veremos que Python conoce diferentes tipos de valores que podemos usar (caracteres, números, etc.). Cada tipo de dato tiene su propia sintaxis por tanto Python debe saber cómo diferenciarlos.

* Quizás esto suene un poco confuso, pero la primera distinción a tomar en cuenta, que además es muy importante:

    * Los nombres de las variables no van entre comillas.
    * Los valores que asignamos a las variables van entre comillas si son cadenas de texto.
    * Se ven muy similares, pero son muy diferentes, y tienen propósitos completamente diferentes.

Veamos el siguiente ejemplo:

```python
cantante = "Taylor Swift"
print("cantante") # Este es una cadena de texto con el valor "cantante"
print(cantante) # Este es una variable con el valor "Taylor Swift"
```

Como observamos imprimir una cadena de texto "cantante" no es lo mismo que imprimir el valor de la variable cantante. En el primer caso, estamos imprimiendo el texto "cantante", mientras que en el segundo caso estamos imprimiendo el valor de la variable cantante que es "Taylor Swift".

Por otro lado, la función **`print()`** permite incluir distintos tipos de datos y variables como argumentos:

```python
# Imprimimos la fecha
print("Hoy es miércoles", 28, "de agosto del", 2024)
```

```python
# Imprimimos la cantidad de cursos

# Primero creamos una variable con el valor (# de cursos)
numero_cursos = 5

# Luego, ejecutamos la función print()
print("Tengo", numero_cursos, "cursos.")
```

Si recordamos un poco de matemáticas, este símbolo se te hará un poco familiar `=`. En matemáticas, el símbolo `=` se utiliza para indicar que dos expresiones tienen el mismo valor.

En programación no es tan diferente, pero lo llamamos **asingación**, donde usamos este mismo símbolo `=`, con el nombre de la variable a su izquierda y el valor que queremos asignar a la derecha.

De esta manera estamos asignando un valor y nombrandolo usando el símbolo de asignación `=`, por tanto luego podemos usar fácilmente el nombre de la variable en lugar de tener que escribir el valor completo cada vez que lo necesitemos.

Pensemos en las variables como cajas donde podemos guardar objetos. Cada caja tiene un nombre (variable) y un contenido (valor). En este caso nuestra variable `x` es una especie de etiqueta que le pusimos a la caja para identificarla. En esta caja ponemos lo que nosotros quisieramos, como algún número, en este caso el valor `20`. Por tanto en Python, el término **variable** se refiere a la etiqueta que le ponemos a la caja, mientras que el término **valor** se refiere al contenido de la caja.

## **2. Nombres de las variables**
Hay que tener en cuenta que los **nombres de variables** `mensaje` y `x` utilizados arriba no son parte de Python.

En principio, podrías usar cualquier nombre que te guste. Incluso si cambias la variable mensaje por algo como `barcelona` o `real_madrid`, el ejemplo seguiría funcionando:

```python
#¿Qué sucede aquí?
CR_7 = 889
print(CR_7 + 1)
```

Sin embargo, los nombres de las variables solo son válidos si cumplen con ciertas reglas. De lo contrario, obtendrás un error de Python.

Aquí hay algunas reglas para nombrar variables en Python:

* Los nombres de las variables deben comenzar con una letra (a - z, A - B) o un guion bajo (_).
* El resto del nombre de la variable puede consistir en letras, números y guiones bajos.

Asimismo, si bien puedes usar cualquier nombre mientras sean válidos, hay ciertos convenciones de nombramiento que es mejor seguir (https://peps.python.org/pep-0008/#naming-conventions).

Por el momento, es suficiente que recuerdes seguir lo siguiente:
* utiliza nombres de variables claros, significativos y descriptivos para que tu código siga siendo comprensible
* emplea el estilo minusculas_con_guiones_bajos, con caracteres en minúscula y guiones bajos para separar palabras
* no utilices nombres reservados, como print o sum (estos se volverán de otro color en los Notebooks). Aquí (https://www.w3schools.com/python/python_ref_keywords.asp) hay una lista de palabras reservadas en Python.

Por ejemplo, el siguiente nombre de variable es válido, mucho más descriptivo que `x` y sigue las convenciones de nomenclatura:

```python
# Creamos una variable que contenga los segundos por semana
segundos_semana = 60 * 60 * 24 * 7
print(segundos_semana)


```

## **3. Tipos de datos en Python**
Python es un lenguaje orientado a objetos(datos). Los tipos de datos más comunes son strings, integers y floats. Cada vez que escribimos un texto en Python, estamos utilizando **strings** (cadena de caracteres). Los números más comunes, por otro lado, son **integers** (por ejemplo, -1, 0, 100) y **floats**,que representan números reales (por ejemplo, 3.14, -42.0).

descarga.jpeg

```python
# Integer (entero)

print(6)
```

```python
# Float (decimal)

print(2.14)
```

```python
#  String (cadena de caracteres)

print("Coldplay")
```

Asimismo, puedes hacer que Python te diga el tipo de dato usando la función incorporada `type()`. Notarás que Python se refiere a los números enteros como **int**, decimales como **float**, y los strings como **str**.

```python
# Tipo de dato: 21

type(21)
```

```python
# Tipo de dato: 3.14

type(3.14)
```

```python
# Tipo de dato: 12.0

type(12.0)
```

```python
# Tipo de dato: "Shawn Mendes"

type("Shawn Mendes")
```

###**3.1. Integers**
Los números enteros pueden ser números negativos o positivos.

```python
# Suma de dos integers
numero_1 = 7
numero_2 = 9

print(numero_1 + numero_2)
print(7 + 9)
```

```python
# Resta de dos integers
print(numero_1 - numero_2)
print(type(numero_1 - numero_2))
```

```python
# Comprueba el tipo de dato de la multiplicación de los resultados anteriores
print(numero_1 * numero_2)
```

### **3.2. Floats**
Floats representan números reales; son un superconjunto de números enteros pero también incluyen "números decimales".

# **4. Conversión de datos**
### **Convertir integers en floats:** Cuando convertimos un integer en un float, en realidad no cambiamos el valor del número.

```python
# Convirtamos 2 en un float

float(2)
```

### **Convertir floats en integers**: Si convertimos un float en un integer, potencialmente podríamos perder algo de información.

```python
# Convirtamos 1.1 en integer, resultará en la pérdida de información.

int(1.1)
```

### **Convertir de strings a integers o floats**: A veces, podemos tener un string que contiene un número dentro de ella. Si este es el caso, podemos convertir ese string que representa un número en un entero usando `int()`:

```python
# Convirtamos un string en un integer

int('1')
```

```python
numero = "20"
convertir = int(numero)
print(type(numero))
print(numero)
print(type(convertir))
print(convertir)
```

```python
# Convirtamos un string en un integer con error

int('1 or 2 people')
```

Ten en cuenta que las cadenas se pueden representar con comillas simples ('1.2') o doble comillas ("1.2"), pero nunca combinados ("1.2').

```python
# Analizamos las siguientes divisiones
print(6/4)
print(type(6/4))
```

```python
# ¿Cuál es la diferencia entre el uso de / y //?
print(6//4)
print(type(6//4))
```

Después, mostrará estos datos y verificará sus tipos usando `type()`.

* Crear 4 variables
* Calcular la tasa de interacción: **tasa_interaccion = (me_gusta + cantidad_comentarios) / interaccion_general**
