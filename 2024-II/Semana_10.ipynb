
#**Base de datos**

Una base de datos es un conjunto organizado de datos que se almacenan y gestionan de manera que se puedan recuperar y utilizar de forma eficiente.

Las bases de datos permiten almacenar grandes cantidades de información de manera estructurada y acceder a ella rápidamente para realizar consultas y análisis.

## **Componentes Principales de una Base de Datos**
1. Datos: La información propiamente dicha que se almacena.
2. Tablas: Estructuras que organizan los datos en filas y columnas. Cada tabla suele representar una entidad específica (por ejemplo, clientes, productos, ventas).
3. Registros (filas): Cada fila de una tabla representa un conjunto individual de datos (valor).
4. Campos (columnas): Cada columna de una tabla representa un tipo de dato específico (por ejemplo, nombre, dirección, precio).

# ¿Excel o CSV (Comma-Separated Values)?

## **Semejanzas**:
1. Almacenamiento de Datos Tabulares: Ambos formatos se utilizan para almacenar datos en forma de filas y columnas.
2. Compatibilidad: Ambos pueden ser importados y exportados desde diversas aplicaciones de software, incluidas bases de datos, herramientas de análisis de datos y programas de estadísticas.
3. Uso Común en Análisis de Datos: Ambos formatos son ampliamente utilizados en el análisis de datos y son compatibles con muchos lenguajes de programación y herramientas de análisis de datos (por ejemplo, Python, R, pandas).

--------------------------

## **Diferencias**
1. **Formato de archivo**:
- Excel: Es un formato binario o basado en XML (XLS o XLSX) que puede contener múltiples hojas, gráficos, fórmulas, macros, y estilos de formato.
- CSV: Es un formato de texto plano donde los valores están separados por comas. No puede contener múltiples hojas ni formato avanzado.

2. **Capacidades y funcionalidades**:
- Excel: Soporta fórmulas, gráficos, tablas dinámicas, y una variedad de funcionalidades avanzadas de análisis de datos.
- CSV: Solo almacena datos sin ningún tipo de formato o funcionalidad adicional. No soporta fórmulas, gráficos ni otros elementos avanzados.

3. **Tamaño del archivo**:
- Excel: Los archivos pueden ser más grandes debido al almacenamiento de formato, fórmulas, y otros elementos adicionales.
- CSV: Generalmente, los archivos CSV son más pequeños y más ligeros, ya que solo contienen datos sin formato.

4. **Compatibilidad con software**:
- Excel: Requiere un software específico (Microsoft Excel) para abrir, visualizar y editar los archivos con todas sus funcionalidades.
- CSV: Puede ser abierto y editado con cualquier editor de texto (como Notepad) o software que soporte archivos de texto plano. Es más universal y fácil de compartir.

5. **Facilidad de lectura y escritura por programas**:
- Excel: Requiere librerías específicas para su manipulación en programación (por ejemplo, openpyxl o xlrd en Python).
- CSV: Es muy fácil de leer y escribir en programación usando herramientas estándar en casi todos los lenguajes de programación (por ejemplo, csv en Python).

# **Gráfico de pastel (Pie charts)**

## Pueden revisar más en el siguiente enlace:
https://matplotlib.org/stable/plot_types/stats/pie.html#sphx-glr-plot-types-stats-pie-py

```python
# Filtrar los partidos donde el Real Madrid es el equipo local o visitante
partidos_real_madrid = df_csv[(df_csv['HomeTeam'] == 'Real Madrid')]

# Inicializar contadores para partidos ganados, perdidos y empatados
ganados = 0
perdidos = 0
empatados = 0

# Contar el número de partidos ganados, perdidos y empatados
for index, row in partidos_real_madrid.iterrows():
    if row['HomeTeam'] == 'Real Madrid':
        if row['FTR'] == 'H':
            ganados = ganados + 1
        elif row['FTR'] == 'A':
            perdidos = perdidos + 1
        else:
            empatados = empatados + 1

# Datos para el gráfico de pastel
labels = ['Ganados', 'Perdidos', 'Empatados']
sizes = [ganados, perdidos, empatados]
colors = ['#4CAF50', '#FF5733', '#FFC300']
explode = (0, 0, 0) ## explode es una tupla que puede ser usada para resaltar una sección del gráfico

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False, startangle=140)
## shadow=False quita la sombra y elgráfico se ve 2d, si deseas efecto shadpw=True y juega con explode
plt.title('Resultados de los Partidos como local del Real Madrid (2018-2019)')
plt.axis('equal')  # asegura que el gráfico de pastel se dibuje como un círculo perfecto

# Guardar el gráfico
plt.savefig('resultados_real_madrid.png', dpi=300)

# Mostrar el gráfico
plt.show()
```

```python
# Calcular el promedio de goles anotados como local para cada equipo
promedio_goles_por_equipo = df_csv.groupby('HomeTeam')['FTHG'].mean()

#Esta parte del código agrupa el DataFrame df_csv por la columna 'HomeTeam'.
# Agrupar por 'HomeTeam' significa que se crearán grupos de filas donde cada grupo
# contiene todos los partidos jugados por el mismo equipo como local.

# Crear el gráfico de barras con pandas
promedio_goles_por_equipo.plot(kind='bar', figsize=(12, 6), color='orange')

# Configurar el título y las etiquetas de los ejes
plt.title('Promedio de Goles Anotados como Local por Equipo')
plt.xlabel('Equipo')
plt.ylabel('Promedio de Goles Anotados')

# Rotar los nombres de los equipos en el eje x para una mejor visualización
plt.xticks(rotation=90, ha='right')

# Mostrar el gráfico
plt.tight_layout()

# Guardar el gráfico
plt.savefig('Promedio_por_equipo_local.png', dpi=300)

plt.show()
```

```python
# Filtrar los partidos donde el Real Madrid es el equipo local
partidos_local_real_madrid = df_csv[df_csv['HomeTeam'] == 'Real Madrid']

# Calcular el promedio de goles anotados como local por el Real Madrid
promedio_goles_local_real_madrid = partidos_local_real_madrid['FTHG'].mean()

# Filtrar los partidos donde el Barcelona es el equipo local
partidos_local_barcelona = df_csv[df_csv['HomeTeam'] == 'Barcelona']

# Calcular el promedio de goles anotados como local por el Barcelona
promedio_goles_local_barcelona = partidos_local_barcelona['FTHG'].mean()

# Promedios de goles anotados como local por el Real Madrid y el Barcelona
promedios = [promedio_goles_local_real_madrid, promedio_goles_local_barcelona]
equipos = ['Real Madrid', 'Barcelona']

# Crear el gráfico de pastel
plt.figure(figsize=(8, 6))
plt.pie(promedios, labels=equipos, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral'])

# Configurar el título
plt.title('Promedio de Goles Anotados como Local\ndel Real Madrid y el Barcelona')

# Mostrar el gráfico
plt.axis('equal')

# Guardar el gráfico
plt.savefig('Promedio_goles_madird_barcelona.png', dpi=300)

plt.show()
```
