## **Ejercicio 1**
Imagina que trabajas en una agencia de publicidad y necesitas calcular el presupuesto de una campaña publicitaria en Facebook, Instagram y Tiktok.

* Crea una variable y asignale un valor a la publicidad en Facebook.
* Crea una variable y asignale un valor a la publicidad en Instagram.
* Crea una variable y asignale un valor a la publicidad en Tiktok. **Recuerda usar int o float**
* Crea una variable y asignale la suma total de los presupuestos de cada plataforma.
* Crea nueva variables donde asignes el porcentaje del presupuesto total por plataforma. **Fórmula: porcentaje_plataforma = (presupuesto_plataforma / total_presupuesto) * 100**
* Usa print para mostrar el presupuesto destinado a cada plataforma y el porcentaje que representa del presupuesto total. **No te olvides comentar tu código**

----
## **Solución**

```python
# En primer lugar, creamos 3 variables que guarden el presupuesto de la campaña publicitaria en 3 redes sociales: Facebook, Instagram y Tiktok
presupuesto_facebook = 1400.14
presupuesto_instagram = 2000
presupuesto_tiktok = 2700.50

# En segundo lugar, creo otra variable donde asigne la suma total de los presupuestos en las redes sociales
total_presupuesto = presupuesto_facebook + presupuesto_instagram + presupuesto_tiktok

# En tercer lugar, creo nuevas variables para asignar el porcentaje del presupuesto total invertido en las redes sociales
porcentaje_facebook = (presupuesto_facebook / total_presupuesto) * 100
porcentaje_ig = (presupuesto_instagram / total_presupuesto) * 100
porcentaje_tiktok = (presupuesto_tiktok / total_presupuesto) * 100

# Finalmente, imprimiré el presupuesto invertido en cada red social y el porcentaje que representa del total del presupuesto invertido
print("En Facebook se invirtió", presupuesto_facebook, "soles. El monto representó el", porcentaje_facebook, "% del presuepuesto total.")
print("En Instagram se invirtió", presupuesto_instagram, "soles. El monto representó el", porcentaje_ig, "% del presuepuesto total.")
print("En Tiktok se invirtió", presupuesto_tiktok, "soles. El monto representó el", porcentaje_tiktok, "% del presuepuesto total.")
```

## **Ejercicio 2**

* Crea una variable llamada marca y asígnale el nombre de una marca.
* Crea una variable llamada slogan y asígnale una frase publicitaria.
* Crea una variable y asígnale el valor de la inversión en televisión como un entero (int).
* Crea una variable y asígnale el valor de la inversión en radio como un número con decimales (float).
* Crea una variable y asígnale el valor de la inversión en medios digitales como un string.
* Luego, convierte el primer presupuesto a un tipo de dato float, el segundo presupuesto a un tipo de dato int, y al último prespuesto a float.
* Crea una variable y asignale la suma de los presupuestos de cada plataforma.
* Finalmente, crea una variable que combine la marca, el slogan y el presupuesto total en una sola cadena de texto.
* No olvides usar la función type() para verificar el tipo de dato de las variables después de la conversión.
* Usa print() para mostrar el mensaje completo, los presupuestos convertidos y los tipos de datos de cada variable.

----
## **Solución**

```python
# Creamos la variable marca para guardar un string (cadena de caracteres)
marca = "Inka Kola"

# Creamos la variable slogan para guardar una frase
slogan = "¡Hagamos un Perú que nos dé gusto!"

# Creamos tres variables para guardar el presupuesto invertido en la campaña publicitaria de la marca "Inka Kola"
inversión_tv = 1500
inversión_radio = 850.30
inversión_medios_digitales = "2700"

# Convertimos los valores de las variables anteriores en nuevas variables
conversión_tv = float(inversión_tv)
conversión_radio = int(inversión_radio)
conversión_medios_digitales = float(inversión_medios_digitales)

# Verificamos los tipos de datos resultantes
print(type(conversión_tv))
print(type(conversión_radio))
print(type(conversión_medios_digitales))

# Imprimimos los valores para cada plataforma
print("Se invirtió", conversión_tv, "soles para la campaña en televisión de Inka Kola.")
print("Se invirtió", conversión_radio, "soles para la campaña en radio de Inka Kola.")
print("Se invirtió", conversión_medios_digitales, "soles para la campaña en medios digitales de Inka Kola.")

# Creamos una variable donde asignamos la suma total de la inversión
suma_presupuesto = conversión_tv + conversión_radio + conversión_medios_digitales

# Imprimimos el mensaje completo
print("La campaña publicitaria de la marca", marca, "con el concepto", slogan, "costó en total", suma_presupuesto, "soles.")



```
