## **.count()**

Esta función te brinda la **cantidad de veces que un valor especificado aparece** en una cadena o en una lista.

```python
# Creamos una variable con un cadena de texto (str)
comentario = "Los que le ponen mala reseña están exagerando demasiado. Es una buena película, tiene su encanto macabro, conserva el estilo de Burton junto con los actores originales, incluso marionetas y animación. Para darle su toque, el humor y bromas a lo Burton están muy bien metidas. Sois unos exagerados, si queréis un copia y pega volved a ver la primera pero no desprecieis la segunda."
print(comentario.count("Burton"))

# Creamos una variable con un lista (list)
desaprobados = [1, 2, 2, 3, 2, 4]
print(desaprobados.count(2))
```

## **len()**
Esta función te ofrece la **longitud (o cantidad de elementos) de un objeto** como una cadena, lista, etc.

```python
# Creamos una variable con un cadena de texto (str)
saludo = "Hola"
print(len(saludo))

# Creamos una variable con un lista (list)
numeros = [1, 2, 3, 4, 5]
print(len(numeros))
```

```python
# Con entero (int) no funciona directamente porque no tiene longitud.
mi_entero = 123
print(len(mi_entero))
```

```python
# Con flotante (float) tampoco es aplicable directamente.
mi_float = 3.14159
print(len(mi_float))
```

## **.startswith()**
Esta función te indica **`True` si la cadena comienza con el valor especificado**; de lo contrario, devuelve `False`.

```python
mensaje = "Hola mundo"
print(mensaje.startswith("Hola"))
print(mensaje.startswith("mundo"))
```

## **.endswith()**

Esta función te indica **`True` si la cadena termina con el valor especificado**; de lo contrario, devuelve `False`.

```python
mensaje_2 = "Hala madrid"
print(mensaje_2.endswith("madrid"))
print(mensaje_2.endswith("Hala"))
```

## **.isalnum()**
Esta función te indica **`True` si todos los caracteres en la cadena son alfanuméricos**; de lo contrario, devuelve `False`. No se aplica a otros tipos de datos como enteros, flotantes, booleanos o listas.

```python
nombre = "Python123"
print(nombre.isalnum())

nombre_2 = "Python 123"
print(nombre_2.isalnum())  # No debe haber un espacio
```

```python
numero_entero = 123
print(numero_entero.isalnum())
```

## **.isalpha()**
Esta función **verifica si todos los caracteres en una cadena son solo letras (sin números ni símbolos)** y devuelve `True` si lo son.

```python
curso = "Python"
print(curso.isalpha())
# nombre = "Python123" , esta variable se creo antes
print(nombre.isalpha())
```

```python
mi_booleano = True
print(mi_booleano.isalpha())
```

## **.isdigit()**
Esta función **verifica si todos los caracteres en una cadena son dígitos (números)** y devuelve `True` si lo son.

```python
nombre_3 = "456"
print(nombre_3.isdigit())

nombre_4 = "45.6"
print(nombre_4.isdigit())
```

```python
# numero_entero = 123, esta variable lo creamos antes
print(numero_entero.isdigit())  # No aplicable a enteros
```
