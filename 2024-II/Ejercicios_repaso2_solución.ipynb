## **Ejercicio 1**
Crear un programa que solicite información al usuario y valide los datos ingresados con expresiones booleanas.

* Solicita al usuario su nombre completo (cadena de caracteres). Usa `input()` para obtener esta información.
* Solicita al usuario su edad. Usa `input()` para obtener esta información y convierte el valor a **int**.
* Solicita al usuario un código promocional (cadena de caracteres) y valida con expresiones booleanas que empiece con "PROMO". Usa `startswith()`.

```python
# Solicitamos el nombre al usuario y lo guardamos en una variable
nombre_usuario = input("Ingrese su nombre completo: ")

### Si nos indicaran corroborar que los datos ingresados en la variable nombre_usuario solo sean caracteres no numéricos
### Podemos usar la función isdigit()

# Verificamos
print("Datos incorrectamente ingresados." * nombre_usuario.isdigit() or "Datos correctamente ingresados.")
```

```python
# Solicitamos la edad del usuario y lo guardamos en una variable
edad_usuario = input("Ingrese su edad: ")

## Recuerda que si bien el usuario ingresa datos numéricos la función input() siempre te devuelve un string
## Por ello, lo convertiremos al tipo de dato con el cual deseo trabajar

# Si nos indicaran que la edad ingresada debe ser mayor de edad podemos crear con expresiones booleanas una condición para verificar

mayor_edad = int(edad_usuario) >= 18

# Verificamos
print("Usted es mayor de edad." * mayor_edad or "Usted no es mayor de edad.")
```

```python
# Solicitamos al usuario un código promocional y lo guardamos en una variable
codigo_promocional = input("Ingrese su código promocional: ")

# Debemos validar el código con expresiones booleanas y que empiece con "PROMO"
verificar_codigo = codigo_promocional.startswith("PROMO") or codigo_promocional.startswith("promo")

# Verificamos
print("Su código promocional ha sido ingresado correctamente." * verificar_codigo or "El código no es válido.")
```

```python
## Si nos indicaran que se debe cumplir con los datos bien ingresado correctamente
## Con expresiones booleanas podemos establecer la verificación

datos_nombres = int(nombre_usuario.isdigit()) == 0

datos_edad = int(mayor_edad) == 1

datos_codigo = int(verificar_codigo) == 1

# Verificamos

print("Datos correctamente ingresados." * (datos_nombres and datos_edad and datos_codigo) or "Datos incorrectos.")
```

## **Ejercicio 2**
Desarrollar un programa que permita a los alumnos crear un perfil básico de un beneficiario de un programa de desarrollo comunitario.

* Se debe brindar el nombre completo y validar que no contenga caracteres numéricos.
* Se debe brindar la edad y validar que sea mayor igual a 18 años.
* Se debe brindar el ingreso mensual y validar que sea menor a 1500 soles.
* Usa `print()` y las expresiones booleanas para verificar que los datos fueron correctamente ingresados.

```python
# Primero, solicitamos el nombre al beneficiario con la función input() y lo guardamos en una variable
nombre_beneficiario = input("Ingrese su nombre completo: ")

# Creamos otra variable para asignarle el valor binario para verificar que el nombre ha sido ingresado correctamente
validar_nombre = int(int(nombre_beneficiario.isdigit()) == 0)

# Segundo, solicitamos la edad del beneficiario y lo guardamos en una variable
edad_beneficiario = int(input("Ingrese su edad: "))

# Creamos otra variable para asignarle el valor binario para verificar que la edad ingresada es mayor e igual a 18
validar_edad = int(edad_beneficiario >= 18)

# Tercero, solicitamos el ingreso mensual del beneficiario y lo guardamos en una variable
ingreso_mensual = int(input("Ingrese su ingreso mensual: "))

# Creamos otra variable para guardar el valor binario para verificar que el ingreso sea menor a 1500
validar_ingreso = int(ingreso_mensual < 1500)

# Verificamos todos los datos ingresados
print("Datos correctamente ingresados." * (validar_nombre and validar_edad and validar_ingreso) or "Datos no válidos.")


```

* Crea una variable pertinente que contenga una lista con los siguientes elementos `Publicidad, Periodismo, Comunicación para el Desarrollo, Comunicación Audivisual`.

```python
# Creamos una variable con los siguientes elementos
facultad_comunicaciones = ["Publicidad", "Periodismo", "Comunicación para el Desarrollo", "Comunicación Audivisual"]

# Creamos otra variable con  los siguientes datos
facultad_humanidades = ["Lingüística", "Literatura", "Filosofía", "Ciencias de la Información", "Humanidades", "Historia", "Arqueología", "Geografía"]

# Con expresiones booleanas verificaremos si la primera lista es más grande que la segunda usando la función len()
lista_1_grande = len(facultad_comunicaciones) > len(facultad_humanidades)

# Imprimimos el resultado
print("La lista facultad_comunicaciones es más grande que facultad_humanidades." * int(lista_1_grande) or "La lista facultad_humanidades es más grande que facultad_comunicaciones.")

# Tomando en cuenta la posición del elemento "Comunicación Audiovisual" en la lista facultad_comunicaciones, imprimimos dicho elemento
print(facultad_comunicaciones[3])

# Verificamos si el elemento "Trabajo Social" pertenece a la lista facultad_humanidades
validar_elemento = "Trabajo Social" in facultad_humanidades
print("El término 'Trabajo Social' pertenece a lista facultad_humanidades." * int(validar_elemento) or "El término 'Trabajo Social' no pertenece a lista facultad_humanidades." )
```
